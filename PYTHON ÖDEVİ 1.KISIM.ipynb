```python
#Titanik verisinde erkek hayatta kalan ve 30 dan fazla ödeyen yolcuların yaşlarının ortlamasını ve standart sapmasını bulan kodu oluşturun.
```

```python
import seaborn as sns
```

```python
titanic=sns.load_dataset("titanic")
```

```python
titanic
```

```python
# Önce erkek yolcuları filtreliyorum.
```

```python
titanic[titanic["sex"]=="male"]
```

```python
# Hayatta kalan yolcuları buluyorum.
```

```python
titanic[(titanic["sex"]=="male") & (titanic["survived"]==1)]
```

```python
# Şimdi de ödediği ücretin 30'dan fazla olması koşulunu sağlatıyorum.
```

```python
titanic[(titanic["sex"]=="male") & (titanic["survived"]==1)&(titanic["fare"]>30)]
```

```python
df=titanic[(titanic["sex"]=="male") & (titanic["survived"]==1)&(titanic["fare"]>30)]
```

```python
df["age"]
```

```python
df["age"].mean()
```

```python
df["age"].std()
```
